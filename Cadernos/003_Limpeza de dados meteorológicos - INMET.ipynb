```python
import os
read_path =  os.path.join(r"..\Dados brutos\INMET","2021","INMET_CO_DF_A001_BRASILIA_01-01-2021_A_31-12-2021.CSV")
with open(read_path, "r+") as file:
    data = file.read()
    data = data.splitlines()
    print(data[0:20])
```

```python
# Sources:
# https://www.geeksforgeeks.org/how-to-search-and-replace-text-in-a-file-in-python/
# https://stackoverflow.com/questions/33743816/how-to-find-a-filename-that-contains-a-given-string
# https://pynative.com/python-search-for-a-string-in-text-files/#:~:text=Open%20a%20file%20in%20a,current%20line%20and%20line%20number.
estacoes_nao_encontradas = []
for estacao in estacoes_unicas:
    found = 0
    years = ["2021", "2022", "2023"]
    for year in years:
        search_path = os.path.join(r"..\Dados brutos\INMET",year)

        for filename in os.listdir(search_path):
            if ("_"+estacao+"_") in filename:        
                found = 1
                # print(f"Conjunto de dados referente à estação {estacao} foi encontrado")
                read_path = os.path.join(r"..\Dados brutos\INMET",year,filename)

                with open(read_path, "r+") as file:
                    data = file.read()
                    data = data.replace(",",".") # Substituindo vírgulas por pontos para não prejudicar a leitura dos arquivos .csv
                    # Ignorando as primeiras 8 linhas
                    data = data.splitlines()
                    boundary = 9
                    data=data[boundary:]
                    data="\n".join(data)
                    data=data+"\n"
                    file.close()

                write_path = os.path.join(r"..\Dados brutos\INMET",estacao+".csv")

                with open(write_path,"a") as file:
                    file.write(data)
                    file.close()    
    
    if found == 0:
        estacoes_nao_encontradas.append(estacao)
        print(f"\tConjunto de dados referente à estação {estacao} não foi encontrado")
```

```python
from datetime import datetime
def formatDateTime(date_time_UTC):
    date_time_UTC = date_time_UTC.split()
    date = date_time_UTC[0]
    hour, minute = date_time_UTC[1][0:2], date_time_UTC[1][2:]
    date_time = date + " " + hour + " " + minute
    return(datetime.strptime(date_time,"%Y/%m/%d %H %M"))
print(formatDateTime("2021/01/01 0030 UTC"))
```

```python
estacoes_unicas.sort()
column_labels = ["Data","Hora UTC","PRECIPITAÇÃO TOTAL HORÁRIO (mm)","PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO HORARIA (mB)","PRESSÃO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)","PRESSÃO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)","RADIACAO GLOBAL (Kj/m²)","TEMPERATURA DO AR - BULBO SECO HORARIA (°C)","TEMPERATURA DO PONTO DE ORVALHO (°C)","TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)","TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)","TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (°C)","TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (°C)","UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)","UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)","UMIDADE RELATIVA DO AR HORARIA (%)","VENTO DIREÇÃO HORARIA (gr) (° (gr))","VENTO RAJADA MAXIMA (m/s)","VENTO VELOCIDADE HORARIA (m/s)"," "]

for estacao in estacoes_unicas:
    read_path = os.path.join(r"..\Dados brutos\INMET", estacao + ".csv")
    df = pd.read_csv(read_path,encoding="latin",sep=";",header=None)
    df.columns=column_labels
    df=df[["Data","Hora UTC","VENTO RAJADA MAXIMA (m/s)","VENTO VELOCIDADE HORARIA (m/s)"]]
    df["DATA E HORA"] = df["Data"]+" "+df["Hora UTC"]
    df["DATA E HORA"] = df["DATA E HORA"].apply(formatDateTime)
    df=df[["DATA E HORA","VENTO RAJADA MAXIMA (m/s)","VENTO VELOCIDADE HORARIA (m/s)"]]
    df.head()
    write_path = os.path.join(r"..\Dados tratados","VENTO - " + estacao + ".csv")
    df.to_csv(write_path, encoding = "utf-8", index=False)
```
